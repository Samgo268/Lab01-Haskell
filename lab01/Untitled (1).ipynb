```haskell
sumaCuadrados :: Int -> Int -> Int
sumaCuadrados x y = (x*x)+(y*y)
sumaCuadrados 5 6
```

```haskell
dobleLista :: [Int] -> [Int]
dobleLista [] = []
dobleLista (x:xs) = (x*2): dobleLista xs
dobleLista [2, 3, 4, 5]
```

```haskell
clasificarNota :: Int -> String
clasificarNota x 
    | x < 3 = "Reprobado"
    | x < 4 = "En recuperacion"
    | 4 <= x  = "Aprobado"
clasificarNota 4
```

```haskell
fibonacci :: Int -> Int
fibonacci 0 = 1
fibonacci 1 = 1
fibonacci x = fibonacci(x-1) + fibonacci(x-2)
fibonacci 2
```
