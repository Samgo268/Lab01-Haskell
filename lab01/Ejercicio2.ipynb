```haskell
head [10,20,30,40]
tail [10,20,30,40]
take 3 [1..10]
drop 4 [1..10]
length [1,2,3,4,5]
```

```haskell
primerElemento :: [a] -> a
primerElemento xs = head xs
```

```haskell
productoLista :: [Int] -> Int
productoLista [] = 1
productoLista xs= head xs * productoLista(tail xs)
productoLista [2, 3, 4, 5]
```

```haskell
eliminarNegativos :: [Int] -> [Int]
eliminarNegativos [] = []
eliminarNegativos (x:xs) = if x < 0 then eliminarNegativos xs else [x] ++ eliminarNegativos xs
eliminarNegativos [3, -2, -1, 3, 4, -5]
```

```haskell
invertir :: [a] -> [a]
invertir [] = []
invertir (x:xs) = invertir xs ++ [x]
invertir [2, 4, 5, 6]
```
